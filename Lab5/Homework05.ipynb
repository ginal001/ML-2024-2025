# Solution

We are going to create synthetic data which is separable by ellipse. To do that, we are creating concentric circles of data. Then, we are going to train SVM model with polynomial kernel (d = 2) and plot the results of our classifier. I mostly use the code from labs.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_circles

# Generate Data (Concentric Circles for Non-Linear Separation)
np.random.seed(42)
X, y = make_circles(n_samples=300, factor=0.5, noise=0.05)

# Convert labels from (0,1) to (-1,1) for SVM
y = np.where(y == 0, -1, 1)

# Train SVM with Polynomial Kernel (Degree 2, c=0)
clf = SVC(kernel='poly', degree=2, C=1.0, coef0=0, gamma=1)
# scikit uses a complex scaled gamma for the kernels. Here we want to use non-optimal (but simplest!) kernels
# K(x, x') = (x^T x')^2
clf.fit(X, y)

# Create Meshgrid for Plotting Decision Boundary
x_min, x_max = X[:, 0].min() - 0.2, X[:, 0].max() + 0.2
y_min, y_max = X[:, 1].min() - 0.2, X[:, 1].max() + 0.2
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))

# Flatten and Stack Grid for Prediction
grid = np.c_[xx.ravel(), yy.ravel()]
decision = clf.decision_function(grid)
decision = decision.reshape(xx.shape)

# Plot Data Points and Decision Boundary
plt.figure(figsize=(10, 7))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolors='k', s=50, alpha=0.7)

# Plot Decision Boundary and Margins
plt.contour(xx, yy, decision, levels=[0], linewidths=2, colors='k')
plt.contour(xx, yy, decision, levels=[-1, 1], linestyles='--', colors='grey')

# Highlight Support Vectors
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=100, linewidth=1, facecolors='none', edgecolors='k', label='Support Vectors')

plt.title('SVM with Degree 2 Polynomial Kernel')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.grid(True)

plt.gca().set_aspect('equal')  #  making the plot square
plt.show()
```

And now, let's plot the result in three-dimensional space. I am adding a two dimensional manifold on which there are our data points. We are going to look at our graph from different perspectives.

```python
def plot_3d(ax, elev=30, azim=45):
    # Visualize Explicit Mapping phi(x) = [x1^2, sqrt(2) x1 x2, x2^2]
    
    # Apply phi(x) to each point
    phi_X = np.array([[x[0]**2, np.sqrt(2)*x[0]*x[1], x[1]**2] for x in X])
    
    # Apply phi(x) to the meshgrid for decision boundary
    phi_grid = np.array([[x[0]**2, np.sqrt(2)*x[0]*x[1], x[1]**2] for x in grid])
    
    # Scatter plot of transformed data
    ax.scatter(phi_X[:, 0], phi_X[:, 1], phi_X[:, 2], c=y, cmap='bwr', edgecolors='k', s=25, alpha=0.7, label='Data Points')
    
    # Plot only the decision boundary contour (i.e. where decision == 0)
    grid_decision = clf.decision_function(grid)
    contour_mask = np.isclose(grid_decision, 0, atol=0.01)  # select points near decision boundary
    ax.scatter(phi_grid[contour_mask, 0], phi_grid[contour_mask, 1], phi_grid[contour_mask, 2],
               c='k', s=5, label='Decision Boundary')
    
    # Drawing a mainfold
    x_range = np.linspace(0, phi_X[:, 0].max()  , 500) 
    z_range = np.linspace(0, phi_X[:, 2].max()  , 500)
    X_cone, Z_cone = np.meshgrid(x_range, z_range)
    Y_cone_positive = np.sqrt(2* X_cone * Z_cone)
    Y_cone_negative = -np.sqrt(2* X_cone * Z_cone)
    ax.plot_surface(X_cone, Y_cone_positive, Z_cone, color='green', alpha=0.3, edgecolor='none')
    ax.plot_surface(X_cone, Y_cone_negative, Z_cone, color='green', alpha=0.3, edgecolor='none')
    
    # Set axis labels and title for 3D plot
    ax.set_title('Explicit Mapping $\phi(x) = [x_1^2, \sqrt{2} x_1 x_2, x_2^2]$ with Decision Boundary')
    ax.set_xlabel('$x_1^2$')
    ax.set_ylabel('$\sqrt{2} x_1 x_2$')
    ax.set_zlabel('$x_2^2$')
    
    # Add legend and grid
    ax.legend()
    ax.grid(True)
    
    ax.view_init(elev=elev, azim=azim) #changing perspective
```

```python
view_angles = [(30, 45), (30, 180), (15, 0), (60, 90)]

fig = plt.figure(figsize=(16, 12))
n_plots = len(view_angles)
ncols = 2
nrows = (n_plots + 1) // ncols

for idx, (elev, azim) in enumerate(view_angles):
    ax = fig.add_subplot(nrows, ncols, idx+1, projection='3d')
    plot_3d(ax, elev=elev, azim=azim)

plt.tight_layout()
plt.show()
```

# Answers for questions
**1.** By using the kernel trick, we avoid computing the entire hyperplane in the original space. Instead, we determine a decision boundary within a two-dimensional space. In essence, we find the intersection between our two-dimensional manifold (illustrated as the green region in the plots that contains our data points) and the hyperplane, which results in an elliptical boundary.

**2.** Our decision boundary in two-dimensional space is elliptical because we are applying non-linear transform from three-dimensions to two-dimensions (which results in non-linear shape). It is a intersection of hyperplane with two-dimensional manifold. To be more precise, we have $$w_1z_1 + w_2z_2 + w_3z_3 + b = 0$$ and we can substitute $$w_1x_1^2 + w_2\sqrt 2 x_1x_2 + w_3x_2^2 + b = 0$$ which is clearly a quadratic equation and explains shape of our decision boundary.
